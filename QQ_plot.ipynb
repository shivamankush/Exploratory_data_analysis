# QQ PLOT

```python
# QQ plot are used to compare any two distributions but in general it is used for normal distributions
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
sns.kdeplot(df['sepal_length'])
```

```python
temp=sorted(df['sepal_length'].tolist())
```

```python
y_quant=[]

for i in range(1,101):
    y_quant.append(np.percentile(temp,i))
```

```python
# mean=0,sd=1,datapoints=1000
samples=np.random.normal(loc=0,scale=1,size=1000)
```

```python
x_quant=[]
for i in range(1,101):
    x_quant.append(np.percentile(samples,i))
```

```python
sns.scatterplot(x=x_quant,y=y_quant)
```

```python
## Using statsmodel
import statsmodels.api as sm
```

```python
# create a QQ plot of the two sets of data
fig=sm.qqplot(df['sepal_length'],line='45',fit=True)

# Add a title and labels to the plot
plt.title('QQ plot')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('sample Quantiles')
# show the plot
plt.show()
```
