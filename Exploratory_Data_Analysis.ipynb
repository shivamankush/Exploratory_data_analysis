## What is KDE here?

#### KDE stands for Kernel Density Estimation. It is a smooth curve that estimates the probability density function (PDF) of a continuous random variable. Simply put, KDE helps you visualize the underlying distribution of a dataset by creating a smooth curve instead of showing discrete bins (like in a histogram)

### Why to use KDE

#### A histogram shows the distribution of data using discrete bins (bar plots). However, a histogram can be sensitive to the number of bins and bin widths, which may result in a choppy or misleading representation of the data. KDE solves this problem by drawing a smooth curve that represents the overall shape of the distribution, making it easier to interpret

##  Handling Correlation and display using Heatmap

#### -->Pandas dataframe.corr() is used to find the pairwise correlation of all columns in the dataframe.
#### -->Any NA values are automatically excluded.
#### -->Here, we are using Pearson correlation coefficient

```python
corr_matrix=df.drop(columns=['Id','Species']).corr(method='pearson')
display(corr_matrix)
```

```python
plt.figure(figsize = (8,6))
sns.heatmap(corr_matrix, annot=True, cmap = 'mako')
plt.title('Correlation Matrix')
plt.show()
```
